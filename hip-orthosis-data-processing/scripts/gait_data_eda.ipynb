```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = 'Arial'
plt.rcParams['font.weight'] = 'bold'
plt.rcParams['axes.labelweight'] = 'bold'
plt.rcParams['axes.titleweight'] = 'bold'
plt.rcParams['axes.linewidth'] = 1.5

MAX_HIP_ANGLE = 50
MIN_HIP_ANKLE = -30
MAX_KNEE_ANGLE = 85
MIN_KNEE_ANKLE = -15
MAX_ANKLE_ANGLE = 25
MIN_ANKLE_ANKLE = -35
ISCONSTRAINED = {'Left': "Left_Constrained", 'Right': "Right_Unconstrained"}


def plot_all_subjects(joint, side, option):
    fig = plt.figure(figsize=(14, 6))
    nrows = 3
    ncols = 6
    subject_levels = ["S{}".format(i) for i in range(1, 18)]
    
    for idx, sub in enumerate(subject_levels):    
        ax = fig.add_subplot(nrows, ncols, idx + 1) 
        condition = (joint, 'NoOrtho', sub, side)
        joint_angle_natural = gait_df.loc[condition].to_numpy()
        joint_velocity_natural = np.gradient(joint_angle_natural)
        
        condition = (joint, 'Ortho', sub, side)
        joint_angle_ortho = gait_df.loc[condition].to_numpy()
        joint_velocity_ortho = np.gradient(joint_angle_ortho)
        
        if option == "phase":
            plot_gait_phase_plane(ax, joint, joint_angle_natural, joint_velocity_natural)
            plot_gait_phase_plane(ax, joint, joint_angle_ortho, joint_velocity_ortho)
        elif option == "spatiotemporal":
            plot_gait_angle(ax, joint, joint_angle_natural)
            plot_gait_angle(ax, joint, joint_angle_ortho)
        else:
            print("Specify a valid name of plot.")
        set_subplot_properties(ax, joint, side, idx, ncols)
                
    ax = fig.add_subplot(nrows, ncols, idx + 2) 
    if option == "phase":
        plot_gait_phase_plane(ax, joint, None, None)
        plot_gait_phase_plane(ax, joint, None, None)
    elif option == "spatiotemporal":
        plot_gait_angle(ax, joint, [])
        plot_gait_angle(ax, joint, [])
    set_subplot_properties(ax, joint, side, idx+1, ncols)
    
    plt.savefig("Constrained_Knee_{j}_{s}.pdf".format(j=joint, s=side), format='pdf', bbox_inches='tight')
    

def set_subplot_properties(ax, joint, side, plot_idx, ncols):
    # Only keep y axis label for first column
    if plot_idx % ncols != 0:
        ax.set_yticklabels([])
    
    # Show only y grid
    ax.yaxis.grid(True)
        
    # Get joint limits
    y_min = 0
    y_max = 0
    joint_limits = {'Hip': (MIN_HIP_ANKLE, MAX_HIP_ANGLE),
                    'Knee': (MIN_KNEE_ANKLE, MAX_KNEE_ANGLE),
                    'Ankle': (MIN_ANKLE_ANKLE, MAX_ANKLE_ANGLE)}
    y_min, y_max = joint_limits[joint]
        
    # Set y limits for all subplots
    ax.set_xlim([0, 100])
    ax.set_ylim([y_min, y_max])  
    
    # Set y grid for all subplots at a inverval of 10
    ax.set_yticks(np.arange(y_min, y_max, 20))  
    
    # Set y label for first column
    if plot_idx % ncols == 0:
        ax.set_ylabel("{} Angle (deg)".format(joint))
        
    # Set x label for last row
    if plot_idx >= (ncols * 2):
        ax.set_xlabel("Gait Cycle (%)")
        
    # Put legend on the last subplot
    if plot_idx == (ncols * 3 - 1):
        ax.legend(['Natural', 'Ortho'])
        # remove all label ticks
        ax.set_xticks([])
        ax.set_yticks([])
        ax.set_xlabel('')
        # remove all labels and label ticks
        ax.set_xticklabels([])
        ax.set_yticklabels([])
        # remove the frame of the chart
        ax.set_frame_on(False)
        # remove grid
        ax.grid(False)
        # Add a text box with information about the joint
        textstr = 'Joint: {j} \nSide: {s}'.format(j=joint, s=ISCONSTRAINED[side])
        # put the text box in the upper left in axes coords
        props = dict(boxstyle='round', facecolor='white', alpha=0.5)
        ax.text(0, y_min+10, textstr, fontweight='bold', bbox=props)
        

def plot_gait_phase_plane(ax, joint, joint_angle, joint_velocity):
    ax.plot(joint_angle, joint_velocity)
    if joint == 'Hip':
        ax.set_ylim([MIN_HIP_ANKLE, MAX_HIP_ANGLE])
    elif joint == 'Knee':
        ax.set_ylim([MIN_KNEE_ANKLE, MAX_KNEE_ANGLE])
    elif joint == 'Ankle':
        ax.set_ylim([MIN_ANKLE_ANKLE, MAX_ANKLE_ANGLE])
    else:
        print("Specify a valid joint.")    
    
def plot_gait_angle(ax, joint, joint_angle):
    ax.plot(range(0, len(joint_angle)), joint_angle)

gait_df = pd.read_pickle("../data/caren_gait_analysis_data/gait_ortho.pkl")

plot_all_subjects(joint='Hip', side='Left', option="spatiotemporal")
plot_all_subjects(joint='Hip', side='Right', option="spatiotemporal")
plot_all_subjects(joint='Knee', side='Left', option="spatiotemporal")
plot_all_subjects(joint='Knee', side='Right', option="spatiotemporal")
plot_all_subjects(joint='Ankle', side='Left', option="spatiotemporal")
plot_all_subjects(joint='Ankle', side='Right', option="spatiotemporal")



```
